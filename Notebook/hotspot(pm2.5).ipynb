```python
import pandas as pd
import plotly.express as px
```

```python
df=pd.read_csv("C:\\Users\\Chinmaya holla\\Desktop\\Airpollution\\Data\\Processed_data\\analysis_ready_data.csv")  
```

```python
coords_data = {
    'Station': [
        'BapujiNagar', 'Hebbal', 'Hombegowda', 'Jayanagar', 'Peenya', 
        'Silkboard', 'BTM', 'Kadabasenahalli', 'RVCE'
    ],
    'lat': [12.9568, 13.0354, 12.9375, 12.9208, 13.0285, 12.9176, 12.9166, 12.9473, 12.9214],
    'lon': [77.5397, 77.5988, 77.5949, 77.5842, 77.5197, 77.6233, 77.6101, 77.6983, 77.5024]
}
df_coords = pd.DataFrame(coords_data)
```

```python
station_stats = df.groupby('Station')[['PM2_5_AQI', 'PM2_5']].mean().reset_index()


map_data = pd.merge(station_stats, df_coords, on='Station')
```

```python

fig = px.scatter_mapbox(
    map_data,
    lat="lat",
    lon="lon",
    color="PM2_5_AQI",          # Color dots based on pollution (Red=High, Yellow=Low)
    size="PM2_5_AQI",           # Bigger dots = More pollution
    hover_name="Station",       # Show station name when hovering
    hover_data={"PM2_5": True, "lat": False, "lon": False}, # Extra info on hover
    color_continuous_scale=px.colors.sequential.YlOrRd, # Yellow -> Orange -> Red Palette
    size_max=30,                # Max size of the dots
    zoom=10.5,
    center={"lat": 12.9716, "lon": 77.5946}, # Center of Bangalore
    mapbox_style="carto-positron", # Clean map style
    title="Bangalore Air Quality Hotspots (Average PM2.5 AQI)"
)

fig.show()
```

The Big Red Dots: These are your "Hotspots" (likely Peenya and Silkboard).

The Small Yellow Dots: These are your cleaner zones (likely RVCE or Kadabasenahalli).

Interaction: Try hovering your mouse over the dots to see the exact AQI numbers.
